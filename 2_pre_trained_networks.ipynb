```python
from torchvision import models
```

```python
dir(models)
```

```python
alexnet = models.AlexNet()
```

```python
resnet = models.resnet101(pretrained=True)
```

```python
resnet
```

```python
from torchvision import transforms
preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )])
```

```python
from PIL import Image
img = Image.open("../../StarTrekPorthos.jpg")
```

```python
img
```

```python
img_t = preprocess(img)
```

```python
import torch
batch_t = torch.unsqueeze(img_t, 0)
```

```python
resnet.eval()
```

```python
out = resnet(batch_t)
out
```

```python
with open('../data/p1ch2/imagenet_classes.txt') as f:
    labels = [line.strip() for line in f.readlines()]
```

```python
_, index = torch.max(out, 1)
```

```python
percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100
labels[index[0]], percentage[index[0]].item()
```

```python
_, indices = torch.sort(out, descending=True)
[(labels[idx], percentage[idx].item()) for idx in indices[0][:5]]
```

```python
from ptflops import get_model_complexity_info
macs, params = get_model_complexity_info(resnet, (3, 224, 224), as_strings=True,
                                           print_per_layer_stat=True, verbose=True)
```

```python
print('{:<30}  {:<8}'.format('Computational complexity: ', macs))
print('{:<30}  {:<8}'.format('Number of parameters: ', params))
```

```python
mobilenetv2 = torch.hub.load('pytorch/vision:v0.10.0', 'mobilenet_v2', pretrained=True)
```

```python
mobilenetv2.eval()
out = mobilenetv2(batch_t)
_, index = torch.max(out, 1)
percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100
labels[index[0]], percentage[index[0]].item()
_, indices = torch.sort(out, descending=True)
[(labels[idx], percentage[idx].item()) for idx in indices[0][:5]]
```

```python
macs, params = get_model_complexity_info(mobilenetv2, (3, 224, 224), as_strings=True,
                                           print_per_layer_stat=True, verbose=True)
```

```python
print('{:<30}  {:<8}'.format('Computational complexity: ', macs))
print('{:<30}  {:<8}'.format('Number of parameters: ', params))
```
